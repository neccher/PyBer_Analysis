```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with as.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create plot and add label for legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# CReate labels for axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set y limit
plt.ylim(0,45)
#Title
plt.title("PyBer Fare by Month")
#Add Legend
plt.legend()
```

```python
#Formatting with object interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
#Set axes and labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
# Y axis limit
ax.set_ylim(0,45)
# Add legen and gridlines
ax.legend()
ax.grid()
```

```python
#Create plot fo Bar Chart
plt.bar(x_axis, y_axis, color="green", label="Boston")
#Axes labes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#title
plt.title("PyBer Fare by Month")
#legend
plt.legend()
```

```python
#Create Horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
#Add labes
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#Create plot with object oriented approach ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#SkillDrill
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()
```

```python
#Scatter plot
plt.plot(x_axis, y_axis, "o")
```

```python
# Anotha one
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,45)
plt.gca().invert_yaxis()
```

```python
#Bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
#Object oriented approacgh
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,c="skyblue", alpha=0.2, s=[i*5 for i in y_axis], edgecolors="black", label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
#Pie
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, colors=colors, autopct="%.1f%%")
plt.show()
```

```python
#Object oriented
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%.1f%%", shadow=True, startangle=90, counterclock=False, colors=colors)
plt.show()
```
